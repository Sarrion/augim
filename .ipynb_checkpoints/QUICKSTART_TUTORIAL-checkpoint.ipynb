This notebook is a quick start guide to use augim, an image augmentation module for python. In order to import the module, you can use:

import sys  
sys.path.append('/path/where/augim/module/is')

And then:

```python
import augim
```

For a quick demonstration you can instantiante the <b>Augmenter()</b> object an use its <b>.augment()</b> method. This way you will have applied the default augment_scheme in the default images. 

```python
augments = augim.Augmenter().augment()
```

We can visualize the augmented images using the <b>augs_display()</b> function

```python
augim.augs_display(augments, original_im_n = 1, range_of_aug_ims = (0,9))
```

We printed the second image (<b>original_im_n</b> parameter) first nine augments (<b>range_of_aug_ims </b>parameter), with its Ids as titles. This titles consists in the name of the original image ('gotobelair') and the name of the augment or augments applied separated by an underscore. 

By default 100 augments are built, you can tune this parameter using the <b>augment_req</b> parameter:

```python
augments = augim.Augmenter(augment_req = 150).augment()
augim.augs_display(augments, original_im_n = 1, range_of_aug_ims = (144,150), no_of_columns = 2)
```

To use your own images you need to instantiate an <b>Input_Data_Formatter()</b> object and pass two parameters to it:
<ul text-align: justify>
    <li><b>sources:</b> A list of strings representing where the images are.</li>
    <li>
        <b>metadata:</b> A list of pandas DataFrames with an Id column with the image names and other parameters needed for augments. Using default augments, the only one that needs extra parameters is crop, which needs XMin, XMax, YMin, Ymax.
    </li>
</ul>
